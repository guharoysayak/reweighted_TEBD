```python
import numpy as np
from pathlib import Path

run_dir = Path("20260201_184137_fermion_test_L6_chi1000_g1.5_T20.0_N250")

data = np.load(run_dir / "results.npz")

print(data.files)
```

```python
# Example run of a L = 6 system and comparing against exact

import matplotlib.pyplot as plt

T = 20
N = 250

plt.plot(np.arange(1,N+1,1)*(T/N),data['ni_psite'][1],label=r'$i=1$ rTEBD, $\gamma=1.5$',color='b')
plt.plot(np.arange(1,N+1,1)*(T/N),data['ni_psite_sch'][1],'^',label=r'$i=1$, Exact',color='b')

plt.plot(np.arange(1,N+1,1)*(T/N),data['ni_psite'][2],label=r'$i=2$ rTEBD, $\gamma=1.5$',color='r')
plt.plot(np.arange(1,N+1,1)*(T/N),data['ni_psite_sch'][2],'^',label=r'$i=2$, Exact',color='r')

plt.plot(np.arange(1,N+1,1)*(T/N),data['ni_psite'][3],label=r'$i=3$ rTEBD, $\gamma=1.5$',color='g')
plt.plot(np.arange(1,N+1,1)*(T/N),data['ni_psite_sch'][3],'^',label=r'$i=3$, Exact',color='g')

plt.plot(np.arange(1,N+1,1)*(T/N),data['ni_psite'][4],label=r'$i=4$ rTEBD, $\gamma=1.5$',color='k')
plt.plot(np.arange(1,N+1,1)*(T/N),data['ni_psite_sch'][4],'^',label=r'$i=4$, Exact',color='k')

plt.xlabel(r'$t$',fontsize=14)
plt.ylabel(r'$\langle n_i \rangle$ for $i=1,2,3,4$',fontsize=14)
plt.legend()
```
